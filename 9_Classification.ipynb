```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
data.drop(columns = ['customerID'], inplace = True)
```

```python
X = data.drop(columns = ['Churn'])
y = data['Churn']
```

```python
oe = OrdinalEncoder()
categorical_cols = X.select_dtypes('object').columns
X[categorical_cols] = oe.fit_transform(X[categorical_cols])
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.svm import SVC
model2 = SVC()
model2.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier()
model3.fit(X_train,y_train)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
print(accuracy_score(y_test,model3.predict(X_test)))
sns.heatmap(confusion_matrix(y_test,model3.predict(X_test)),cmap= 'Blues',annot = True)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
print(accuracy_score(y_test,model2.predict(X_test)))
sns.heatmap(confusion_matrix(y_test,model2.predict(X_test)),cmap= 'Blues',annot = True)
```
